```python
%matplotlib inline
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv.imread('../images/video-x/frame24.jpg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
mean_c = cv.adaptiveThreshold(img_gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 19, 6)
plt.imshow(mean_c, cmap='gray')
```

```python
blur = cv.blur(img, (5,5))
plt.imshow(blur)
```

```python
gaus = cv.GaussianBlur(img, (5,5), 0)
plt.imshow(gaus)
```

```python
median = cv.medianBlur(img, 5)
plt.imshow(median)
```

```python
_,mask = cv.threshold(img, 250, 255, cv.THRESH_BINARY_INV)
kernal = np.ones([5,5], np.uint8)
dilation = cv.dilate(mask, kernal)
plt.imshow(dilation)
```

```python
blur = cv.GaussianBlur(img_gray, (11,11), 0)
canny = cv.Canny(blur, 100, 150)
plt.imshow(canny, cmap='gray')
```
