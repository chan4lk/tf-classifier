```python
def show_image(img):
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    
def show_gimage(img):
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap='gray')
```

```python
img = cv.imread('../images/video-x/frame24.jpg', cv.IMREAD_GRAYSCALE)
show_gimage(img)
```

```python
_, thresh = cv.threshold(img, 7, 255, cv.THRESH_BINARY)
show_gimage(thresh)
```

```python
_, thresh_inv = cv.threshold(img, 5, 255, cv.THRESH_BINARY_INV)
show_gimage(thresh_inv)
```

```python
_, thresh_trunch = cv.threshold(img, 5, 255, cv.THRESH_TRUNC)
show_gimage(thresh_trunch)
```

```python
cv.imshow('HelloWorld', img)
cv.waitKey(0) 
cv.destroyAllWindows()
```

```python
def nothing(x):
    pass

cv.namedWindow("image")
cv.createTrackbar("Threshold image", "image", 128, 255, nothing)

while True:
    value = cv.getTrackbarPos("Threshold image", "image")
    _, thresh_trunch = cv.threshold(img, value, 255, cv.THRESH_BINARY)
    _, thresh_inv = cv.threshold(img, value, 255, cv.THRESH_BINARY_INV)
    
    cv.imshow("image", img)
    cv.imshow("th bin", thresh_trunch)    
    cv.imshow("th inv", thresh_inv)

    
    key = cv.waitKey(100)
    if key == 27:
        break
cv.destroyAllWindows()   
```
